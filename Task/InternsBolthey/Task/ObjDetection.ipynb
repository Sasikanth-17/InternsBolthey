```python
import cv2 as cv
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def display_ims(image1,image2,title1="Original",title2="Detected"):

    plt.figure(figsize=(20,10))

    plt.subplot(1, 3, 1)
    plt.imshow(image1)
    plt.title(title1)
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(image2)
    plt.title(title2)
    plt.axis('off')
```

```python
image_path=r"C:\Users\Azista Industries\OneDrive\Desktop\Task\Images\cars.jpg"
template_path=r"C:\Users\Azista Industries\OneDrive\Desktop\Task\Images\maintemplate.jpg"
#template_path=r"C:\Users\Azista Industries\OneDrive\Desktop\Task\Images\temp.jpg"
#template_path=r"C:\Users\Azista Industries\OneDrive\Desktop\Task\Images\templ.jpg"
image=cv.imread(image_path)
image_gray=cv.cvtColor(image,cv.COLOR_BGR2GRAY)
template=cv.imread(template_path,0)
```

```python
plt.imshow(image_gray,cmap='gray')
plt.axis('off')
```

```python
w,h=template.shape[1],template.shape[0]
```

Using A Template


```python
result=cv.matchTemplate(image_gray,template,cv.TM_CCOEFF_NORMED)

```

```python
threshold=0.5
loc=np.where(result>=threshold)
for i in zip(*loc[::-1]):
    # cv.rectangle(image,i,(i[0]+w,i[1]+h),(0,255,0),1)
    cv.rectangle(image,i,(i[0]+w,i[1]+h),(0,255,0),thickness=1)
display_ims(template,image)
```

```python
cv.imshow('Detected',image)
cv.waitKey(0)
cv.destroyAllWindows()
```
